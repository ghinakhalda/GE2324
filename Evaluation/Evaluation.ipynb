```python
from bs4 import BeautifulSoup
import re
import pandas as pd
from evaluate import load
import matplotlib.pyplot as plt
```

```python
dfResult = pd.read_csv('test_result.csv')
dfResult
```

# Perplexity

```python
print(f"Mean: {dfResult['perplexity'].mean()}")
print(f"Median: {dfResult['perplexity'].median()}")
print(f"Variance: {dfResult['perplexity'].var()}")
print(f"Maximum: {dfResult['perplexity'].max()}")
print(f"Minimum: {dfResult['perplexity'].min()}")
```

```python
plt.figure(figsize=(8, 5))
plt.boxplot(dfResult['perplexity'],patch_artist=True, vert=False)
plt.title('The Distribution of Perplexity', weight='bold')
plt.xlabel('Perplexity Score')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

# Plagiarized trigrams

```python
print(f"Mean: {dfResult['plagiarized_trigrams'].mean()}")
print(f"Median: {dfResult['plagiarized_trigrams'].median()}")
print(f"Variance: {dfResult['plagiarized_trigrams'].var()}")
print(f"Maximum: {dfResult['plagiarized_trigrams'].max()}")
print(f"Minimum: {dfResult['plagiarized_trigrams'].min()}")
```

```python
plt.plot(dfResult['plagiarized_trigrams'])
plt.title('Percentages of Plagiarized trigrams over each result')
plt.xlabel('Result')
plt.ylabel('% Plagiarized Trigrams')
plt.show()
```
